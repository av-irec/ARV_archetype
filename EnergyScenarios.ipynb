```python
Msplit = [1, 1, 1, 1, 0, 0, 0, 0]
PV = [1, 1, 0, 0, 1, 1, 0, 0]
BC = [1, 0, 1, 0, 1, 0, 1, 0]

args = [PE_ht, PE_cl, PE_lig, PE_dev, PE_dhw, PE_ht_rt, PE_cl_rt, PE_lig_rt, PE_dev_rt, PE_dhw_rt]

def preMedium(variable):
        
    output = {}
    for i in variable:
        aux = []
        for j in variable[i]:
            aux.append(sum(variable[i][str(j)]))
        output.setdefault(i, mean(aux))                
    return output        
```

```python
for i in range(0,len(args)):
    args[i] = preMedium(args[i])
PE_ht, PE_cl, PE_lig, PE_dev, PE_dhw, PE_ht_rt, PE_cl_rt, PE_lig_rt, PE_dev_rt, PE_dhw_rt = args
```

```python
PE_ht
```

```python
output_scenarios = {}
for k in PE_ht_rt:
    output_scenarios.setdefault(k, {})
    for i in range(0, len(Msplit)):
        key = []
        value = []
        if Msplit[i] == 1:
            key.append('Split_rf+')
            value.append(PE_cl_rt[k] + PE_ht_rt[k])
        else:
            key.append('Split_nrf+')
            value.append(PE_cl[k] + PE_ht[k])
        if PV[i] == 1:
            key.append('PV_yes+')
            if Msplit[i] == 0 and BC[i] == 0:
                value.append(-PV_estimation[k]['Split_nrf+PV_yes+BC_no']*2.968) # En este caso al ser ARV he añadido el coeficiente de paso de electricidad de Baleares, habría que buscar una forma de añadir como input el de península evitando IFs
            elif Msplit[i] == 0 and BC[i] == 1:
                value.append(-PV_estimation[k]['Split_nrf+PV_yes+BC_yes']*2.968)
            elif Msplit[i] == 1 and BC[i] == 0:
                value.append(-PV_estimation[k]['Split_rf+PV_yes+BC_no']*2.968)
            elif Msplit[i] == 1 and BC[i] == 1:
                value.append(-PV_estimation[k]['Split_rf+PV_yes+BC_yes']*2.968)
        else:
            key.append('PV_no+')
            value.append(0)
        if BC[i] == 1:
            key.append('BC_yes')
            value.append(PE_dhw_rt[k])

        else:
            key.append('BC_no')
            value.append(PE_dhw[k])


        output_scenarios[k].setdefault(("".join(key)), sum(value))
```
