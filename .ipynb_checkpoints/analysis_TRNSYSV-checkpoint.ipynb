```python
#import libraries
from repositories import csvTRNSYS
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
from utils import comfortValues, comfortGraph, transformation_EP_CO2
```

```python
#execute readTRSNSYS
n_zones = 1
name = 'FE35'
n_viv = 1
typeStep = csvTRNSYS.HourlyMulti() 
output = csvTRNSYS.readTRNSYS(typeStep, n_viv, name)
output_reader, P_ht, P_cl, P_lig, P_dev, P_dhw = output.start()
```

```python
# Read heating demand
P_htd = pd.DataFrame(P_ht)
P_htd
```

```python
#transform to Dataframe
heating_demand = pd.concat([P_htd[col].explode().reset_index(drop=True) for col in P_htd.columns],axis=1)
heating_demand
```

```python
# create index
start_date = datetime.datetime(year=2023, month=1, day=1, hour=0, minute=0, second=0)
end_date = datetime.datetime(year=2023, month=12, day=31, hour=23, minute=0, second=0)

# Generate a sequence of hourly datetime objects for the year
hourly_datetime_range = pd.date_range(start=start_date, end=end_date, freq='H')

# Create a DataFrame with the hourly datetime objects as the index
heating_demand.index = hourly_datetime_range
heating_demand
```

```python
columns = sorted(heating_demand.columns)
columns
```

```python
#investigate types
df2 = pd.DataFrame([value.split("_") for value in columns], columns=["Retrofit_type","Orientation","Wall","Roof"])

for idx, row in df2.iterrows():
    if row["Retrofit_type"] != 'BC0':
        df2.at[idx, "Type"] = row["Retrofit_type"][0]
        df2.at[idx, "Windows"] = row["Retrofit_type"][1]
        df2.at[idx, "Ecological"] = row["Retrofit_type"][2]
        df2.at[idx, "Material"] = row["Retrofit_type"][3]
    else:
        df2.at[idx, "Type"] = 'B'
        df2.at[idx, "Windows"] = None
        df2.at[idx, "Ecological"] = None
        df2.at[idx, "Material"] = None


# Set the index to the original columns
df2.index = columns

# Display the resulting DataFrame
df2
```

```python
# count typologies
count = [pd.DataFrame(df2[col].value_counts()) for col in df2.columns]
print(count)
```

```python
# try dynamic select

import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

# Sample DataFrame (Replace this with your actual data)
# Assume 'heating_demand' is your DataFrame


heating_demand2 = heating_demand.iloc[:, :-1]
# Initialize the Dash app
app = dash.Dash(__name__)

# Define the app layout
app.layout = html.Div([
    html.H1("Rolling Annual Average for Heating Demand"),
    
    # Dropdown to select filter
    dcc.Dropdown(
        id='filter-selector',
        options=[
            {'label': 'Retrofit_type', 'value': 'Retrofit_type'},
            {'label': 'Orientation', 'value': 'Orientation'},
            {'label': 'Wall', 'value': 'Wall'},
            {'label': 'Roof', 'value': 'Roof'},
        ],
        value='Retrofit_type',  # Default filter
    ),
    
    # Dropdown to select sub-columns
    dcc.Dropdown(
        id='subcolumn-selector',
        multi=True  # Allow multiple sub-column selection
    ),
    
    # Plotly Graph to display rolling average
    dcc.Graph(id='rolling-average-plot'),
])

# Callback to update the sub-column dropdown based on the selected filter
@app.callback(
    Output('subcolumn-selector', 'options'),
    Input('filter-selector', 'value')
)
def update_subcolumn_options(selected_filter):
    # Extract columns based on filter pattern
    filter_options = {
        'Retrofit_type': ['P111', 'P211', 'P121', 'P122', 'P221', 'P222', 'BC0', 'P112', 'P212'],
        'Orientation': ['0', '180', '270', '90'],
        'Wall': ['W8', 'W6'],
        'Roof': ['R10', 'R5', 'R12', 'R14', 'R6', 'R8'],
    }
    
    subcolumns = filter_options.get(selected_filter, [])
    subcolumn_options = [{'label': col, 'value': col} for col in subcolumns]
    
    return subcolumn_options

# Callback to update the graph based on filter and sub-column selection
@app.callback(
    Output('rolling-average-plot', 'figure'),
    Input('filter-selector', 'value'),
    Input('subcolumn-selector', 'value')
)
def update_graph(selected_filter, selected_subcolumns):
    # Combine selected filter and sub-columns to create column names
    selected_columns = [f"{selected_filter}_{subcolumn}" for subcolumn in selected_subcolumns]
    
    # Calculate rolling average for selected columns
    rolling_avg = heating_demand2[selected_columns].rolling(window=window_size).mean()

    # Create a Plotly figure
    fig = px.line(rolling_avg, x=rolling_avg.index, y=selected_columns, title="Rolling Annual Average")
    
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)

```
