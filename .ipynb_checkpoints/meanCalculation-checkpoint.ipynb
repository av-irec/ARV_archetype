#### recorrem la funció calculation a mainv2

Se utiliza **kwargs que nos permite separar en modo clave-valor para que la media se haga independientemente de:
            1) Número de viviendas
            2) Número de simulaciones
            3) Nombre de las carpetas     

```python
def calculation(**kwargs):
    output = {}
    count_args = 0
    for i in kwargs:
        output.setdefault(str(i), {})
        for j in kwargs[i]:
            output[str(i)].setdefault(j, {})
            aux = 0
            count = 0
            for k in kwargs[i][j]:
                aux += sum(kwargs[i][j][k])
                count += 1
            output[str(i)][j] = aux / count

    return output['PE_ht'], output['PE_cl'], output['PE_lig'], output['PE_dev'], output['PE_dhw']

input_data = {
    'PE_ht': PE_ht,  # Replace with actual data for PE_ht
    'PE_cl': PE_cl,  # Replace with actual data for PE_cl
    'PE_lig': PE_lig,  # Replace with actual data for PE_lig
    'PE_dev': PE_dev,  # Replace with actual data for PE_dev
    'PE_dhw': PE_dhw  # Replace with actual data for PE_dhw
}

PE_ht_result, PE_cl_result, PE_lig_result, PE_dev_result, PE_dhw_result = calculation(**input_data)
```
